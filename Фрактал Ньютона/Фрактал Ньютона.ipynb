```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
f = lambda z: z**4 - 1
diff_f = lambda z: 4 * z**3
h = 1
iter_func = lambda z: z - f(z)/diff_f(z) * h 
a, b, c, d = -2, 2, -2, 2
l, r = 640, 480
x = [a + i * (b - a)/l for i in range(l)]
y = [c + i * (d - c)/r for i in range(r)]
points = [(a, b) for a in x for b in y if a != 0 and b!= 0]
```

```python
d = {}
bad_points = [[], []]
for point in points:
    p = point[0] + point[1] * 1j
    for i in range(100):
        p = iter_func(p)
        if f(p) == 0:
            if p not in d:
                d[p] = [[], []]
            d[p][0].append(point[0])
            d[p][1].append(point[1])
            break
    if f(p) != 0:
        bad_points[0].append(point[0])
        bad_points[1].append(point[1])
```

```python
fig, ax = plt.subplots()
i = 0
for key in d:
    ax.scatter(d[key][0], d[key][1], color=f"C{i}", s = 0.1)
    i += 1
ax.scatter(bad_points[0], bad_points[1], color="black", s = 0.5)
ax.set_title("Фрактал Ньютона")
ax.set_xlim(-2, 2)
ax.set_ylim(-2, 2)
ax.set_xlabel("Re(z)")
ax.set_ylabel("Im(z)")
plt.show()
```
